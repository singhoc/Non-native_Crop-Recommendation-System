```python
import requests
import pandas as pd
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows

API_KEYS = ['7NWDTMCRU2DQ8V2VYTVVTFT6B', 'QWFC33L7G5GVCL972KMKA8349', 'BKQD8R6JLWS67AT8D3R4YKLT9', 'L9WG92FVA69ED3NQJ949KD4V2', 'WC52MYN72HBTF6FETJ9JD4K8N', 'YZC8JVZAYSJ4XH3U6L72CAJG5', '6MSHYWNSUKP7T33RJGMA96FBR', '8ZUGURCKMUW4UVGFJ3MT35HP6', 'HJ5WAB9Y4WX8MFXY3CEPC2J9H', '5KE6ZKWHB3SU82G3X54HD7VPZ', 'DDFXCLKA7RNEKS9GGPULM6NU2', 'SMV6Q7DARWHJUXCY2CJXP97QS', 'XC6BJZMYAMR8D4S2KPT3WHQ9Y', 'WFMGSEL7RYG3K8MEWE4J3HJZM', 'P25WHRYF3PMFU4LK2FAMRDY99', 'YLVVTNCG4XZQD8WMLBCDK62BA', '4BLJMPG398SEDNBVTQG7RDZZX', 'BYEX2TCMA5RMDHT3LQC4RB7GK', 'X2L5TQ3CMC3RMWU3CLPQ6J88T', 'T5PZGKH6QSZCJMQFACSKYV6ZS', 'RHRXGSQ3HY6G47KQKSPLZZWEV', 'LEEQWMV79LF25DAJK22G49ES3', 'ANCDFLATFNF9LPJUWTXBXZJER', 'KN966VW7VVSS4Z8EKCB5BL374', 'EH82DCVXFVSZYZD37J43ASXT8']

def fetch_weather_data(district, year, start_month, end_month):
    for api_key in API_KEYS:
         # Adjusting end month if it is less than start month
        nextyear = year
        if end_month < start_month:
            nextyear = str(int(year) + 1)
        start_date = f'{year}-{start_month}-01'
        end_date = f'{nextyear}-{end_month}-30'
        if end_month == '02':
          end_date = f'{nextyear}-{end_month}-28'
        url = f'https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{district}/{start_date}/{end_date}?unitGroup=metric&include=days&key={api_key}&contentType=xlsx'

        response = requests.get(url)
        if response.status_code == 200:
            return response.content
        else:
            print(f"Failed to fetch data for {start_date} to {end_date}. Retrying with next API key.")
    print("Failed to fetch data. All API keys exhausted.")
    return None

def process_weather_data(content):
    if content is None:
        return None

    df = pd.read_excel(content)
    # Selecting only four columns
    relevant_columns = ['temp', 'dew', 'humidity', 'precip','cloudcover','solarradiation','solarenergy','uvindex',]
    processed_data = df[relevant_columns]

    return processed_data

def save_to_excel(data, filename):
    wb = Workbook()
    ws = wb.active

    # Convert DataFrame to rows and add to worksheet
    for r in dataframe_to_rows(data, index=False, header=True):
        ws.append(r)

    # Add default style to the workbook
    for row in ws:
        for cell in row:
            cell.style = 'Normal'

    # Save the workbook
    wb.save(filename)

def main():
    # Read inputs from Excel file
    input_data = pd.read_excel('fetchcustom.xlsx')

    for index, row in input_data.iterrows():
        district = row['district']
        year = row['year']
        crop = row['crop']
        start_month = row['start']
        end_month = row['end']



        start_month_str = str(start_month).zfill(2)  # Zero padding
        end_month_str = str(end_month).zfill(2)      # Zero padding

        filename = f"{district}_{year}_{crop}_{start_month_str}-{end_month_str}_weather.xlsx"

        weather_data = fetch_weather_data(district, year, start_month_str, end_month_str)
        processed_data = process_weather_data(weather_data)

        if processed_data is not None:
            save_to_excel(processed_data, filename)
            print(f"Data saved to {filename}")
        else:
            print("No data available")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd

def select_top_districts(input_file, output_file):
    # Read the Excel file
    df = pd.read_excel(input_file)
    selected_districts = []

    # Iterate over each crop column
    crops = ['rice', 'wheat', 'jowar','bajra','maize','ragi','barley','gram','arhar', 'urad','horsegram','masoor','moong','moth','cowpea','peaandbeans','khesari','groundnut']  # Add more crops here if needed
    for crop in crops:
        # Drop rows with NaN values in the current crop column
        df[f'total_{crop}'] = pd.to_numeric(df[f'total_{crop}'], errors='coerce')
        df_crop = df.dropna(subset=[f'total_{crop}'])

        # Select top 3 districts with highest production for each state and crop
        top_districts = df_crop.groupby('state').apply(lambda x: x.nlargest(1, f'total_{crop}')).reset_index(drop=True)
        # Add crop name to the DataFrame
        top_districts['crop'] = crop
        # Append selected districts to the main DataFrame
        if df_crop is not None and not df_crop.empty:
          # Select top 3 districts with highest production for each state and crop
          top_districts = df_crop.groupby('state').apply(lambda x: x.nlargest(1, f'total_{crop}')).reset_index(drop=True)
          # Add crop name to the DataFrame
          top_districts['crop'] = crop
          # Append selected districts to the main list
          selected_districts.append(top_districts)

    selected_districts_df = pd.concat(selected_districts, ignore_index=True)
    # Save the selected districts to a new Excel file
    selected_districts_df.to_excel(output_file, index=False)

    print(f"Selected districts saved to {output_file}")

# Input and output file paths
input_file = "sample.xlsx"
output_file = "selected_districts.xlsx"

# Call the function
select_top_districts(input_file, output_file)
```

```python
import pandas as pd

def remove_duplicate_rows(file1, file2, output_file):
    # Read the Excel files
    df1 = pd.read_excel(file1)
    df2 = pd.read_excel(file2)

    # Merge the two DataFrames on the columns 'state', 'district', 'year', and 'crop'
    merged_df = pd.merge(df1, df2, on=['state', 'district', 'year', 'crop'], how='left', indicator=True)

    # Select rows from df1 that are not present in df2
    unique_rows_df1 = merged_df[merged_df['_merge'] == 'left_only'].drop(columns='_merge')

    # Save the unique rows to a new Excel file
    unique_rows_df1.to_excel(output_file, index=False)

    print(f"Unique rows saved to {output_file}")

# Input file paths
file1 = "selected_districts.xlsx"
file2 = "selected_districts final.xlsx"
output_file = "unique_rows.xlsx"

# Call the function
remove_duplicate_rows(file1, file2, output_file)
```

```python
import os
import pandas as pd

# Folder containing Excel files
folder_path = "my_data"

# Iterate through each file in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith(".xlsx"):  # Check if the file is an Excel file
        file_path = os.path.join(folder_path, file_name)

        # Load the Excel file into a DataFrame
        df = pd.read_excel(file_path)

        # Check for missing values
        missing_values = df.isnull().sum()
        # Display columns with missing values
        columns_with_missing_values = missing_values[missing_values > 0]
        if len(df) == 0:
          os.remove(file_path)
        elif columns_with_missing_values.empty:
            print("No missing values found.")
        else:
            # Filter out rows with missing values
            df_no_missing = df.dropna()

            # Check if df_no_missing has at least 50 rows and no missing values
            if len(df_no_missing) >= 50:
                # Save filtered DataFrame to the same file
                # Remove the file
                os.remove(file_path)
                print(f"Excel file '{file_name}' removed.")
                df_no_missing.to_excel(file_path, index=False)
                print(f"Filtered Excel file '{file_name}' saved.")
            else:
                # Remove the file
                os.remove(file_path)
                print(f"Excel file '{file_name}' removed.")
```

```python
import pandas as pd

# Load the Excel file into a pandas DataFrame
df = pd.read_excel('sample.xlsx')

# Define the columns to search for maximum values
columns_to_search = ['total_rice', 'total_wheat', 'total_jowar', 'total_bajra', 'total_maize',
                     'total_ragi', 'total_barley', 'total_gram', 'total_arhar', 'total_urad']

max_info = {}  # Dictionary to store max value information

# Iterate over each column
for col in columns_to_search:
    try:
        # Convert column to numeric type and handle errors
        df[col] = df[col].str.replace('.', '0')
        df[col] = pd.to_numeric(df[col])
        max_value = df[col].max()  # Find the maximum value in the column
        max_row = df[df[col] == max_value]  # Find the row where the maximum value occurs
        max_info[col] = {'max_value': max_value, 'row_info': max_row}
    except ValueError as e:
        print(f"Error processing column '{col}': {e}")

# Print the information
for col, info in max_info.items():
    print(f"Maximum value in column '{col}': {info['max_value']}")
    print("Row information:")
    print(info['row_info'])
    print("\n")
```
